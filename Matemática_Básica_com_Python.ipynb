# *Calculadora Numérica*

```python
# Importando bibliotecas necessárias
import numpy as np
import math

# Função para cálculos matemáticos básicos
def calculos_basicos():
    print("=== Calculadora Básica ===")
    a = float(input("Digite o primeiro número: "))
    b = float(input("Digite o segundo número: "))

    print(f"Soma: {a + b}")
    print(f"Subtração: {a - b}")
    print(f"Multiplicação: {a * b}")
    print(f"Divisão: {a / b if b != 0 else 'Erro: Divisão por zero'}")
    print(f"Potência ({a} elevado a {b}): {math.pow(a, b)}")
    print(f"Raiz quadrada de {a}: {math.sqrt(a) if a >= 0 else 'Erro: Raiz de número negativo'}")

# Função para cálculos trigonométricos
def calculos_trigonometricos():
    print("=== Calculadora Trigonométrica ===")
    angulo = float(input("Digite o ângulo em graus: "))
    rad = math.radians(angulo)  # Converte para radianos

    print(f"Seno de {angulo}°: {math.sin(rad):.4f}")
    print(f"Cosseno de {angulo}°: {math.cos(rad):.4f}")
    print(f"Tangente de {angulo}°: {math.tan(rad):.4f}")
    print(f"Arco seno de {math.sin(rad)}: {math.degrees(math.asin(math.sin(rad))):.4f}°")

# Função para cálculos estatísticos
def calculos_estatisticos():
    print("=== Calculadora Estatística ===")
    dados = input("Digite os números separados por espaço: ")
    lista = [float(x) for x in dados.split()]

    print(f"Média: {np.mean(lista):.4f}")
    print(f"Mediana: {np.median(lista):.4f}")
    print(f"Desvio padrão: {np.std(lista):.4f}")
    print(f"Máximo: {np.max(lista)}")
    print(f"Mínimo: {np.min(lista)}")

# Função para operações com matrizes
def operacoes_matrizes():
    print("=== Calculadora de Matrizes ===")
    print("Digite os elementos da matriz 2x2 (A):")
    a11 = float(input("A[1,1]: "))
    a12 = float(input("A[1,2]: "))
    a21 = float(input("A[2,1]: "))
    a22 = float(input("A[2,2]: "))
    matriz_a = np.array([[a11, a12], [a21, a22]])

    print("\nDigite os elementos da matriz 2x2 (B):")
    b11 = float(input("B[1,1]: "))
    b12 = float(input("B[1,2]: "))
    b21 = float(input("B[2,1]: "))
    b22 = float(input("B[2,2]: "))
    matriz_b = np.array([[b11, b12], [b21, b22]])

    print("\nMatriz A:")
    print(matriz_a)
    print("\nMatriz B:")
    print(matriz_b)
    print("\nSoma (A + B):")
    print(matriz_a + matriz_b)
    print("\nMultiplicação (A * B):")
    print(np.dot(matriz_a, matriz_b))
    print("\nDeterminante de A:")
    print(np.linalg.det(matriz_a))
    if np.linalg.det(matriz_a) != 0:
        print("\nInversa de A:")
        print(np.linalg.inv(matriz_a))

# Menu principal
def main():
    while True:
        print("\n=== Calculadora Matemática ===")
        print("1. Cálculos Básicos")
        print("2. Cálculos Trigonométricos")
        print("3. Cálculos Estatísticos")
        print("4. Operações com Matrizes")
        print("5. Sair")

        escolha = input("Escolha uma opção (1-5): ")

        if escolha == '1':
            calculos_basicos()
        elif escolha == '2':
            calculos_trigonometricos()
        elif escolha == '3':
            calculos_estatisticos()
        elif escolha == '4':
            operacoes_matrizes()
        elif escolha == '5':
            print("Saindo...")
            break
        else:
            print("Opção inválida! Tente novamente.")

# Executa o programa
if __name__ == "__main__":
    main()
```
